# DATA MANIPULATION

```python
import pandas as pd
import numpy as np

# 1. Load the pristine clean dataset
df = pd.read_csv('/content/WA_Fn-UseC_-HR-Employee-Attrition.csv')

# 2. Inject missing values (NaNs) into key numerical columns
df.loc[df.sample(frac=0.05).index, 'Age'] = np.nan
df.loc[df.sample(frac=0.04).index, 'TotalWorkingYears'] = np.nan

# 3. Create structural duplicate rows
duplicates = df.sample(n=45, random_state=42)
df = pd.concat([df, duplicates], ignore_index=True)

# 4. Corrupt categorical columns with typos and casing issues
def mess_up_departments(dept):
    rand = np.random.rand()
    if rand < 0.05: return 'sales '        # Trailing space
    if rand < 0.10: return 'SALES'         # Upper casing
    if rand < 0.15: return 'R&D'           # Inconsistent abbreviation
    return dept

df['Department'] = df['Department'].apply(mess_up_departments)

# 5. Turn a clean integer column into an annoying string object
df['MonthlyIncome'] = df['MonthlyIncome'].map(lambda x: f"${x:,.2f}" if np.random.rand() < 0.20 else str(x))

# 6. Inject logical anomalies / outliers
df.loc[df.sample(n=15).index, 'DistanceFromHome'] = -999
df.loc[df.sample(n=10).index, 'YearsAtCompany'] = 999

# Save your new "dirty" dataset
df.to_csv('IBM_HR_Attrition_Dirty.csv', index=False)
print("Dirty dataset created successfully!")
```
